```python
from __future__ import print_function
__author__ = 'yliu'

import pylab as pl
import numpy as np
%pylab inline
```

```python
# Choose six different distributions
distributions = {'gaus':{},'pois':{},'chisq':{}, 'exp':{}, 'gamma':{}, 'binomial':{}}
```

```python
# Generate the sample size for 100 samples
mysize = np.random.choice((range(10,2000)),100,replace=False) 
mysize
```

```python
# Set the parameters for the distributions
mymean = 100
sigma = 2
shape = 100
scale = 1

# Generate samples for each sample size of 6 different distributions
for n in mysize:
    distributions['gaus'][n] = np.random.normal(mymean,sigma,n)
    distributions['pois'][n] = np.random.poisson(mymean,n)
    distributions['chisq'][n] = np.random.chisquare(mymean,size=n)
    distributions['exp'][n] = np.random.exponential(shape,n)
    distributions['gamma'][n] = np.random.gamma(shape,scale,n)
    distributions['binomial'][n] = np.random.binomial(1000,0.1,n)
```

```python
means = {'gaus':{},'pois':{},'chisq':{}, 'exp':{}, 'gamma':{}, 'binomial':{}}

fig_mu_N = pl.figure(figsize=(10,10))

axgaus_mu_n = fig_mu_N.add_subplot(3,2,1)
for nn in distributions['gaus'].iterkeys():
    means['gaus'][nn] = distributions['gaus'][nn].mean()
    axgaus_mu_n.plot(nn, means['gaus'][nn], 'o')
    axgaus_mu_n.set_xlabel('sample size')
    axgaus_mu_n.set_ylabel('sample mean')
    axgaus_mu_n.plot([min(mysize), max(mysize)],[mymean,mymean], 'k')

axpois_mu_n = fig_mu_N.add_subplot(3,2,2)
for nn in distributions['pois'].iterkeys():
    means['pois'][nn] = distributions['pois'][nn].mean()
    axpois_mu_n.plot(nn, means['pois'][nn], 'o')
    axpois_mu_n.set_xlabel('sample size')
    axpois_mu_n.set_ylabel('sample mean')
    axpois_mu_n.plot([min(mysize), max(mysize)],[mymean,mymean], 'k')
    
axchisq_mu_n = fig_mu_N.add_subplot(3,2,3)
for nn in distributions['chisq'].iterkeys():
    means['chisq'][nn] = distributions['chisq'][nn].mean()
    axchisq_mu_n.plot(nn, means['chisq'][nn], 'o')
    axchisq_mu_n.set_xlabel('sample size')
    axchisq_mu_n.set_ylabel('sample mean')
    axchisq_mu_n.plot([min(mysize), max(mysize)],[mymean,mymean], 'k')

axexp_mu_n = fig_mu_N.add_subplot(3,2,4)
for nn in distributions['exp'].iterkeys():
    means['exp'][nn] = distributions['exp'][nn].mean()
    axexp_mu_n.plot(nn, means['exp'][nn], 'o')
    axexp_mu_n.set_xlabel('sample size')
    axexp_mu_n.set_ylabel('sample mean')
    axexp_mu_n.plot([min(mysize), max(mysize)],[shape,shape], 'k')
    
axgamma_mu_n = fig_mu_N.add_subplot(3,2,5)
for nn in distributions['gamma'].iterkeys():
    means['gamma'][nn] = distributions['gamma'][nn].mean()
    axgamma_mu_n.plot(nn, means['gamma'][nn], 'o')
    axgamma_mu_n.set_xlabel('sample size')
    axgamma_mu_n.set_ylabel('sample mean')
    axgamma_mu_n.plot([min(mysize), max(mysize)],[shape,shape], 'k')
    
axbinomial_mu_n = fig_mu_N.add_subplot(3,2,6)
for nn in distributions['binomial'].iterkeys():
    means['binomial'][nn] = distributions['binomial'][nn].mean()
    axbinomial_mu_n.plot(nn, means['binomial'][nn], 'o')
    axbinomial_mu_n.set_xlabel('sample size')
    axbinomial_mu_n.set_ylabel('sample mean')
    axbinomial_mu_n.plot([min(mysize), max(mysize)],[mymean,mymean], 'k')
```

# I expect that the larger the sample size, the closer the sample mean is to the population mean. This is because according to central limit theorem, if the sample size is large enough, the computed values of the average would be approximately distributed according to the normal distribution.

```python
# Generate histogram for all the means
allmeans= []
pl.figure(figsize=(10,10))

# means = {'gaus':{},'pois':{},'chisq':{}, 'exp':{}, 'gamma':{}, 'binomial':{}}

for m in means.iterkeys(): 
    for n in means[m]:
        allmeans.append(means[m][n])
pl.hist(allmeans,bins=50)
pl.xlabel('sample mean')
pl.ylabel('N')
```
